```python
# ============================================================================
# STEP 1: IMPORT LIBRARIES
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime

# Suppress warnings
warnings.filterwarnings('ignore')

# Configure pandas display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

# Set visualization style
sns.set_style('whitegrid')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10

print("✅ Libraries imported successfully!")
print(f"📅 Analysis Date: {datetime.now().strftime('%B %d, %Y')}")
```

## 📥 STEP 2: DATA LOADING & CLEANING

```python
# Load the cleaned dataset
df = pd.read_csv('../data/processed/superstore_clean.csv', 
                 parse_dates=['Order Date', 'Ship Date'])

print(f"✅ Dataset loaded successfully!")
print(f"📊 Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"📅 Date Range: {df['Order Date'].min().date()} to {df['Order Date'].max().date()}")
```

## 📊 STEP 3: KEY BUSINESS METRICS

Calculating the essential KPIs that drive business decisions.

```python
# Calculate key performance indicators
print("\n" + "="*90)
print(" " * 30 + "📈 KEY BUSINESS METRICS")
print("="*90)

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()
total_products = df['Product ID'].nunique()
avg_profit_margin = df['Profit_Margin'].mean()
avg_order_value = df.groupby('Order ID')['Sales'].sum().mean()

print(f"\n💰 Total Sales Revenue:        ${total_sales:,.2f}")
print(f"💵 Total Profit:               ${total_profit:,.2f}")
print(f"📊 Overall Profit Margin:      {avg_profit_margin:.2f}%")
print(f"🛒 Total Orders:               {total_orders:,}")
print(f"👥 Total Customers:            {total_customers:,}")
print(f"📦 Total Products Sold:        {total_products:,}")
print(f"💳 Average Order Value:        ${avg_order_value:,.2f}")
print(f"🔁 Orders per Customer:        {total_orders/total_customers:.2f}")

print("\n" + "="*90)
```

## 📈 STEP 4: SALES & REVENUE ANALYSIS

Analyzing sales trends across different time periods to identify growth patterns.

```python
# Monthly sales trends
monthly_sales = df.groupby('Year_Month').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).reset_index()

# Create visualization
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Monthly sales trend
axes[0].plot(monthly_sales['Year_Month'], monthly_sales['Sales'], 
             marker='o', linewidth=2, markersize=6, color='#2E86AB')
axes[0].set_title('📊 Monthly Sales Trend (2014-2017)', fontsize=16, fontweight='bold', pad=20)
axes[0].set_xlabel('Month', fontsize=12)
axes[0].set_ylabel('Sales ($)', fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Monthly profit trend
axes[1].plot(monthly_sales['Year_Month'], monthly_sales['Profit'], 
             marker='s', linewidth=2, markersize=6, color='#06A77D')
axes[1].set_title('💰 Monthly Profit Trend (2014-2017)', fontsize=16, fontweight='bold', pad=20)
axes[1].set_xlabel('Month', fontsize=12)
axes[1].set_ylabel('Profit ($)', fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print(f"\n✅ Generated monthly sales and profit trends")
```

```python
# Yearly performance analysis
yearly_performance = df.groupby('Year').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).round(2)

yearly_performance.columns = ['Total Sales ($)', 'Total Profit ($)', 'Orders', 'Customers']

# Calculate year-over-year growth
yearly_performance['Sales Growth %'] = yearly_performance['Total Sales ($)'].pct_change() * 100
yearly_performance['Profit Growth %'] = yearly_performance['Total Profit ($)'].pct_change() * 100

print("\n📊 YEARLY PERFORMANCE SUMMARY")
print("=" * 100)
print(yearly_performance)
print("=" * 100)
```

---
## 💰 STEP 5: PROFITABILITY ANALYSIS

Critical analysis of profit margins by category and identification of loss-making products.

```python
# Category profitability analysis
category_profit = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2)

category_profit.columns = ['Total Sales ($)', 'Total Profit ($)', 'Avg Profit Margin (%)']
category_profit = category_profit.sort_values('Total Profit ($)', ascending=False)

print("\n💰 CATEGORY PROFITABILITY ANALYSIS")
print("=" * 80)
print(category_profit)
print("=" * 80)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Profit by Category
axes[0].barh(category_profit.index, category_profit['Total Profit ($)'], 
             color=['green' if x > 0 else 'red' for x in category_profit['Total Profit ($)']])
axes[0].set_title('💵 Total Profit by Category', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Profit ($)', fontsize=12)
axes[0].grid(axis='x', alpha=0.3)

# Plot 2: Profit Margin by Category
axes[1].barh(category_profit.index, category_profit['Avg Profit Margin (%)'], 
             color=['#2E86AB', '#06A77D', '#F18F01'])
axes[1].set_title('📊 Average Profit Margin by Category (%)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Profit Margin (%)', fontsize=12)
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Identify loss-making categories
loss_categories = category_profit[category_profit['Total Profit ($)'] < 0]
if len(loss_categories) > 0:
    print(f"\n⚠️ WARNING: {len(loss_categories)} categories are losing money!")
    print(loss_categories)
```

```python
# Sub-category profitability analysis
subcategory_profit = df.groupby(['Category', 'Sub-Category']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2)

subcategory_profit.columns = ['Sales ($)', 'Profit ($)', 'Margin (%)']
subcategory_profit = subcategory_profit.sort_values('Profit ($)', ascending=True)

print("\n📦 SUB-CATEGORY PROFITABILITY ANALYSIS")
print("=" * 100)
print(subcategory_profit.head(10))
print("\n" + "=" * 100)
print(subcategory_profit.tail(10))
print("=" * 100)

# Identify loss-making sub-categories
loss_subcategories = subcategory_profit[subcategory_profit['Profit ($)'] < 0]
print(f"\n🚨 CRITICAL: {len(loss_subcategories)} sub-categories are LOSING MONEY!")
print("\n❌ LOSS-MAKING SUB-CATEGORIES:")
print("=" * 100)
print(loss_subcategories.sort_values('Profit ($)'))
print("=" * 100)
```

```python
# Discount impact analysis
discount_bins = [0, 0.1, 0.2, 0.3, 1.0]
discount_labels = ['0-10%', '10-20%', '20-30%', '>30%']
df['Discount_Range'] = pd.cut(df['Discount'], bins=discount_bins, labels=discount_labels)

discount_impact = df.groupby('Discount_Range').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean',
    'Order ID': 'count'
}).round(2)

discount_impact.columns = ['Total Sales ($)', 'Total Profit ($)', 'Avg Margin (%)', 'Order Count']

print("\n🎯 DISCOUNT IMPACT ANALYSIS")
print("=" * 100)
print(discount_impact)
print("=" * 100)

# Visualize discount impact
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(discount_impact.index))
width = 0.35

bars1 = ax.bar(x - width/2, discount_impact['Total Sales ($)'], width, label='Sales', color='#2E86AB')
bars2 = ax.bar(x + width/2, discount_impact['Total Profit ($)'], width, label='Profit', color='#06A77D')

ax.set_xlabel('Discount Range', fontsize=12)
ax.set_ylabel('Amount ($)', fontsize=12)
ax.set_title('💸 Impact of Discounts on Sales & Profit', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(discount_impact.index)
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Identify problematic discount levels
high_discount_loss = discount_impact[discount_impact['Avg Margin (%)'] < 5]
if len(high_discount_loss) > 0:
    print(f"\n⚠️ WARNING: Discount ranges with margins < 5%:")
    print(high_discount_loss)
```

---
## 🗺️ STEP 6: REGIONAL PERFORMANCE ANALYSIS

Analyzing sales and profit performance across regions, states, and cities.

```python
# Regional performance analysis
regional_performance = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).round(2)

regional_performance.columns = ['Sales ($)', 'Profit ($)', 'Margin (%)', 'Orders', 'Customers']
regional_performance['Market Share %'] = (regional_performance['Sales ($)'] / regional_performance['Sales ($)'].sum() * 100).round(2)
regional_performance = regional_performance.sort_values('Sales ($)', ascending=False)

print("\n🗺️ REGIONAL PERFORMANCE SUMMARY")
print("=" * 120)
print(regional_performance)
print("=" * 120)

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Sales by Region
axes[0, 0].bar(regional_performance.index, regional_performance['Sales ($)'], color='#2E86AB')
axes[0, 0].set_title('📊 Total Sales by Region', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Region', fontsize=12)
axes[0, 0].set_ylabel('Sales ($)', fontsize=12)
axes[0, 0].grid(axis='y', alpha=0.3)

# Plot 2: Profit by Region
axes[0, 1].bar(regional_performance.index, regional_performance['Profit ($)'], color='#06A77D')
axes[0, 1].set_title('💰 Total Profit by Region', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Region', fontsize=12)
axes[0, 1].set_ylabel('Profit ($)', fontsize=12)
axes[0, 1].grid(axis='y', alpha=0.3)

# Plot 3: Market Share
axes[1, 0].pie(regional_performance['Market Share %'], labels=regional_performance.index, 
               autopct='%1.1f%%', startangle=90, colors=['#2E86AB', '#F18F01', '#06A77D', '#C73E1D'])
axes[1, 0].set_title('🥧 Market Share by Region', fontsize=14, fontweight='bold')

# Plot 4: Profit Margin Comparison
axes[1, 1].barh(regional_performance.index, regional_performance['Margin (%)'], color='#F18F01')
axes[1, 1].set_title('📈 Profit Margin by Region (%)', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Profit Margin (%)', fontsize=12)
axes[1, 1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Identify loss-making regions
loss_regions = regional_performance[regional_performance['Profit ($)'] < 0]
if len(loss_regions) > 0:
    print(f"\n⚠️ WARNING: {len(loss_regions)} regions are losing money!")
    print(loss_regions)
```

```python
# Top and bottom states by profit
state_performance = df.groupby('State').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2)

state_performance.columns = ['Sales ($)', 'Profit ($)', 'Margin (%)']
state_performance = state_performance.sort_values('Profit ($)', ascending=False)

print("\n🏆 TOP 10 STATES BY PROFIT")
print("=" * 80)
print(state_performance.head(10))
print("=" * 80)

print("\n❌ BOTTOM 10 STATES BY PROFIT (Loss-Making)")
print("=" * 80)
print(state_performance.tail(10))
print("=" * 80)

# Identify loss-making states
loss_states = state_performance[state_performance['Profit ($)'] < 0]
print(f"\n⚠️ CRITICAL: {len(loss_states)} states are LOSING MONEY!")
print(f"Total Loss from these states: ${loss_states['Profit ($)'].sum():,.2f}")
```

## 📦 STEP 7: PRODUCT ANALYSIS

Identifying top-performing products and loss-making items that require action.

```python
# Top 10 products by sales
top_products_sales = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).sort_values('Sales', ascending=False).head(10).round(2)

top_products_sales.columns = ['Total Sales ($)', 'Total Profit ($)', 'Quantity Sold']

print("\n🏆 TOP 10 PRODUCTS BY SALES")
print("=" * 100)
print(top_products_sales)
print("=" * 100)

# Visualization
fig, ax = plt.subplots(figsize=(14, 8))
top_products_sales['Total Sales ($)'].plot(kind='barh', ax=ax, color='#2E86AB')
ax.set_title('🏆 Top 10 Products by Sales Revenue', fontsize=16, fontweight='bold')
ax.set_xlabel('Sales ($)', fontsize=12)
ax.set_ylabel('Product Name', fontsize=12)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Loss-making products analysis
product_performance = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).round(2)

product_performance.columns = ['Sales ($)', 'Profit ($)', 'Quantity']

# Identify loss-making products
loss_making_products = product_performance[product_performance['Profit ($)'] < 0].sort_values('Profit ($)')

print(f"\n🚨 CRITICAL ALERT: {len(loss_making_products)} PRODUCTS ARE LOSING MONEY!")
print(f"💸 Total Loss: ${loss_making_products['Profit ($)'].sum():,.2f}")
print("\n❌ TOP 15 LOSS-MAKING PRODUCTS (Urgent Action Required)")
print("=" * 120)
print(loss_making_products.head(15))
print("=" * 120)

# Visualization of worst performers
fig, ax = plt.subplots(figsize=(14, 8))
loss_making_products.head(15)['Profit ($)'].plot(kind='barh', ax=ax, color='#C73E1D')
ax.set_title('❌ Top 15 Loss-Making Products (Urgent Review Required)', fontsize=16, fontweight='bold')
ax.set_xlabel('Loss ($)', fontsize=12)
ax.set_ylabel('Product Name', fontsize=12)
ax.axvline(x=0, color='black', linestyle='--', linewidth=1)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 👥 STEP 8: CUSTOMER ANALYSIS

Analyzing customer segments and identifying top customers for retention strategies.

```python
# Customer segment analysis
segment_analysis = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).round(2)

segment_analysis.columns = ['Total Sales ($)', 'Total Profit ($)', 'Orders', 'Customers']
segment_analysis['Avg Order Value ($)'] = (segment_analysis['Total Sales ($)'] / segment_analysis['Orders']).round(2)
segment_analysis = segment_analysis.sort_values('Total Sales ($)', ascending=False)

print("\n👥 CUSTOMER SEGMENT ANALYSIS")
print("=" * 120)
print(segment_analysis)
print("=" * 120)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Sales by Segment
axes[0].bar(segment_analysis.index, segment_analysis['Total Sales ($)'], color=['#2E86AB', '#F18F01', '#06A77D'])
axes[0].set_title('📊 Total Sales by Customer Segment', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Segment', fontsize=12)
axes[0].set_ylabel('Sales ($)', fontsize=12)
axes[0].grid(axis='y', alpha=0.3)

# Plot 2: Profit by Segment
axes[1].bar(segment_analysis.index, segment_analysis['Total Profit ($)'], color=['#2E86AB', '#F18F01', '#06A77D'])
axes[1].set_title('💰 Total Profit by Customer Segment', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Segment', fontsize=12)
axes[1].set_ylabel('Profit ($)', fontsize=12)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Top 10 customers by profit
customer_performance = df.groupby('Customer Name').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).sort_values('Profit', ascending=False).head(10).round(2)

customer_performance.columns = ['Total Sales ($)', 'Total Profit ($)', 'Order Count']

print("\n🏆 TOP 10 CUSTOMERS BY PROFIT")
print("=" * 100)
print(customer_performance)
print("=" * 100)

# Visualization
fig, ax = plt.subplots(figsize=(14, 8))
customer_performance['Total Profit ($)'].plot(kind='barh', ax=ax, color='#06A77D')
ax.set_title('🏆 Top 10 Most Profitable Customers', fontsize=16, fontweight='bold')
ax.set_xlabel('Profit ($)', fontsize=12)
ax.set_ylabel('Customer Name', fontsize=12)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 INSIGHT: Top 10 customers contribute ${customer_performance['Total Profit ($)'].sum():,.2f} in profit")
```
